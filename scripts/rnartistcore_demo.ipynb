# Introduction

To be able run this notebook efficiently, you need:

* to install [Docker Desktop](https://www.docker.com/products/docker-desktop) on your computer, run it and install the container [fjossinet/rnartistcore](https://hub.docker.com/r/fjossinet/rnartistcore) by typing: ```docker pull fjossinet/rnartistcore```
* In the directory containing this notebook, you need to download and save the script [rnartistcore_docker.sh](https://raw.githubusercontent.com/fjossinet/RNArtistCore/master/scripts/rnartistcore_docker.sh). Make it executable

### RNA 2D described from scratch in the script file

````python
s = '''rnartist {
  file = "/docker/demo1.svg"
  ss {
    rna {
      sequence = "CAACAUCAUACGUACUGCGCCCAAGCGUAACGCGAACACCACGAGUGGUGACUGGUGCUUG"
    }
    bracket_notation =
      "(((..(((..(((..(((((....)))))..)))..(((((....)))))..)))...)))"
  }
  theme {
    details_lvl = 5

    color {
      type="A"
      value = "#A0ECF5"
    }

    color {
      type="a"
      value = "black"
    }

    color {
      type="U"
      value = "#9157E5"
    }

    color {
      type="G"
      value = "darkgreen"
    }

    color {
      type="C"
      value = "#E557E5"
    }

  }
}
'''
````

````python
%store s >demo1.kts
````

````python
# we ask rnartistcore to draw the 2D as described in the script file
````

````python
!./rnartistcore_docker.sh $PWD/demo1.kts
````

````python
#we display the 2D saved in an SVG file
from IPython.core.display import SVG #<-- this line only needs to be run once per notebook
# the output file name is like: filename defined in the script + molecular chain name (default name is 'A'). 
SVG(filename='demo1_A.svg')
````

### RNA 2D inferred from a 3D structure described in a PDB file (PDBID: 1GID)

Download and save the [PDB file 1GID](https://files.rcsb.org/download/1GID.pdb) in your directory before to run the cells below.

````python
!curl "https://files.rcsb.org/download/1GID.pdb" > 1gid.pdb
````

````python
s = '''rnartist {
  file = "/docker/demo3.svg"
  ss {
      pdb {
        name = "B"
        file = "/docker/1gid.pdb"
      }
  }
  theme {
    details_lvl = 5

    color {
      type="A"
      value = "#A0ECF5"
    }

    color {
      type="a"
      value = "black"
    }

    color {
      type="U"
      value = "#9157E5"
    }

    color {
      type="G"
      value = "darkgreen"
    }

    color {
      type="C"
      value = "#E557E5"
    }

  }
}
'''
````

````python
%store s >demo3.kts
````

````python
!./rnartistcore_docker.sh $PWD/demo3.kts
````

````python
SVG(filename='demo3_B.svg')
````
